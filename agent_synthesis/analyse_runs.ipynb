```python
import json
import pandas as pd
```

```python
file_path = r"C:\Users\harsh\Documents\crypto-LLM-exps\AgentBench-Results\v0.2-dev\gpt-4-0613-65\gpt-4-0613\os-dev\runs.jsonl"
df = pd.DataFrame()
all_data = []
with open(file_path, 'r') as file:
    for line in file:
        data = json.loads(line)
        tmp_df = pd.concat([pd.DataFrame([data['output']['result']['result']] , columns=["result"]) , pd.DataFrame(data['output']['history'])], axis=1)
        df = pd.concat([df, tmp_df], axis=1)
        all_data.append(data)
```

```python
df = df.T
```

```python
df.to_csv("runs_analysis/metallama2_synthetic_agentinstruct-axolotl-os-runs-11.5.csv")
```
